```python
import pandas as pd
import ast
```

```python
df = pd.read_csv("/content/tmdb_5000_movies.csv")
df.head()
```

```python
movies = df[['title', 'genres', 'popularity', 'vote_average']]
movies.head()
```

```python
def clean_genres(x):
    genres = []
    for i in ast.literal_eval(x):
        genres.append(i['name'])
    return " ".join(genres)

movies['genres'] = movies['genres'].apply(clean_genres)
movies.head()
```

```python
top_rated = movies.sort_values(by='vote_average', ascending=False).head(10)
top_rated
```

```python
popular_movies = movies.sort_values(by='popularity', ascending=False).head(10)
popular_movies
```

```python
def recommend_by_genre(genre):
    result = movies[movies['genres'].str.contains(genre, case=False, na=False)]
    return result.sort_values(by='vote_average', ascending=False).head(10)
```

```python
genre_input = input("Enter a genre (Action, Comedy, Drama, Romance, etc.): ")
recommend_by_genre(genre_input)
```
